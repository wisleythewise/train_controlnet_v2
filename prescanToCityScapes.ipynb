```python
# Get an image from a file and change the colors of the image 
from PIL import Image

# Open the image file


# Define a function to map the colors of the image
def map_rgb(r, g, b):
    t = 80 

    neural = (68,1,84)
    
    # Sky

    is_sky = 128 - t < r < 128 + t and \
            4 - t < g < 4 + t and \
            252 - t < b < 252+ t 
    if is_sky:
        return  neural

    # Car
    is_car = 255 - t < r < 255 + t and \
            255 - t < g < 255 + t and \
            255 - t < b < 255+ t 
    
    if is_car:
        return  (71, 192, 110)

    # Truck
    is_truck = 1 - t < r < 1 + t and \
            255 - t < g < 255 + t and \
            1 - t < b < 1+ t 
    
    if is_truck:
        return  228, 227, 24


    # Road markings
    is_road_markings = 255 - t < r < 255 + t and \
            255 - t < g < 255 + t and \
            0 - t < b < 0+ t 
    
    if is_road_markings:
        return  253,231,36
    
    # Human
    is_human = 0 - t < r < 0 + t and \
            0 - t < g < 0 + t and \
            255 - t < b < 255+ t 
    
    if is_human:
        return  71, 30, 112

    # Road
    is_road = 126 - t < r < 126 + t and \
            126 - t < g < 126 + t and \
            126 - t < b < 126+ t 
    
    if is_road:
        return  253,231,36



    # Return the new colors
    return neural

# Get all teh images in segmentation folder
import os
path = "/mnt/e/13_Jasper_diffused_samples/complete_dataset/prescan_normal_output/new_scenes_for_new_model/scene-0005/seg/"
files = os.listdir(path)


for i, file in enumerate(files):
    im = Image.open(path + file)

    # Get the size of the image
    width, height = im.size

    # Process every pixel
    for x in range(width):
        for y in range(height):
            # Get the color of the pixel
            r, g, b = im.getpixel((x, y))

            # Change the color of the pixel
            r, g, b = map_rgb(r, g, b)
            # Save the changed pixel
            im.putpixel((x, y), (r, g, b))

    # Save the changed image
    im.save(f"/mnt/e/13_Jasper_diffused_samples/complete_dataset/prescan_normal_output/new_scenes_for_new_model/scene-0005/seg_mapped/{i}.png")
```

```python
from PIL import Image
import numpy as np
import os

def map_rgb(img_array):
    t = 80
    # Using the sky color as the neutral color
    sky_color = np.array([68, 1, 84])

    # Sky
    sky_mask = ((128 - t < img_array[:, :, 0]) & (img_array[:, :, 0] < 128 + t) &
                (4 - t < img_array[:, :, 1]) & (img_array[:, :, 1] < 4 + t) &
                (252 - t < img_array[:, :, 2]) & (img_array[:, :, 2] < 252 + t))
    img_array[sky_mask] = sky_color

    # Other color mappings
    colors_and_conditions = [
        ((255, 255, 255), (71, 192, 110)),  # Car
        ((1, 255, 1), (228, 227, 24)),      # Truck
        ((255, 255, 0), (253, 231, 36)),    # Road markings
        ((0, 0, 255), (71, 30, 112)),       # Human
        ((126, 126, 126), (253, 231, 36))   # Road
    ]

    for original, new_color in colors_and_conditions:
        r, g, b = original
        mask = ((r - t < img_array[:, :, 0]) & (img_array[:, :, 0] < r + t) &
                (g - t < img_array[:, :, 1]) & (img_array[:, :, 1] < g + t) &
                (b - t < img_array[:, :, 2]) & (img_array[:, :, 2] < b + t))
        img_array[mask] = new_color

    # Apply neutral sky color to any regions not specifically mapped
    # This effectively makes the sky color the default for any unmapped colors
    default_mask = np.ones(img_array.shape[:-1], dtype=bool)
    for _, new_color in colors_and_conditions:
        r, g, b = new_color
        color_mask = (img_array[:, :, 0] == r) & (img_array[:, :, 1] == g) & (img_array[:, :, 2] == b)
        default_mask &= ~color_mask
    img_array[default_mask] = sky_color

    return img_array

# Path to the images
path = "/mnt/e/13_Jasper_diffused_samples/complete_dataset/prescan_normal_output/new_scenes_for_new_model/scene-0005/seg/"
output_path = "/mnt/e/13_Jasper_diffused_samples/complete_dataset/prescan_normal_output/new_scenes_for_new_model/scene-0005/seg_mapped/"
os.makedirs(output_path, exist_ok=True)

files = os.listdir(path)
for i, file in enumerate(files):
    # Open and convert the image to an array
    im = Image.open(os.path.join(path, file))
    img_array = np.array(im)

    # Map the RGB values using the function
    new_img_array = map_rgb(img_array)

    # Convert the array back to an image
    new_im = Image.fromarray(new_img_array)

    # Save the changed image
    new_im.save(os.path.join(output_path, f"{i}.png"))
```
